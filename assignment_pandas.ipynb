**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python

high_income = data[data['salary'] == '>50K']['age']
print("Зарплата >50K:")
print(f'Средний возраст: {round(high_income.mean(), 2)}')
print(f'Среднеквадратичное отклонение: {round(high_income.std(), 2)}')

low_income = data[data['salary'] == '<=50K']['age']
print("Зарплата <=50K:")
print(f'Средний возраст: {round(low_income.mean(), 2)}')
print(f'Среднеквадратичное отклонение: {round(low_income.std(), 2)}')
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
men = data[data['sex'] == 'Male']
status= men['marital-status'].str.startswith('Married')
is_married = men[status]
is_single = men[~status]

married_over_50K = is_married[is_married['salary'] == '>50K'].shape[0]


unmarried_over_50K = is_single[is_single['salary'] == '>50K'].shape[0]

married_share = (married_over_50K / len(is_married)) * 100
single_share = (unmarried_over_50K / len(is_single)) * 100

print(f"Доля женатых: {married_share:.2f}")
print(f"Доля холостых: {single_share:.2f}")

```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data["hours-per-week"].max()
print(f'max_hours pr week: {max_hours}')

workaholics = data[data["hours-per-week"] == max_hours]
count = workaholics.shape[0]
print(f"The number of emploees working {max_hours} hours is {count} people.")

rich_percentage = (workaholics['salary'] == '>50K').mean() * 100
print(f'The percentage of rich workaholics is {rich_percentage:.2f} %')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
high_income = data[data['salary'] == '>50K']
low_income = data[data['salary'] == '<=50K']

rich_hours_by_country = high_income.groupby('native-country')['hours-per-week'].mean()
poor_hourse_by_country = low_income.groupby('native-country')['hours-per-week'].mean()

print(f'The mean working hours value for rich employees is {round(rich_hours_by_country,2)}')
print(f'And for poor - {round(poor_hourse_by_country, 2)}')
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
def get_age_group(age):
    if 16 <= age <= 35:
        return 'young'
    elif 35 < age <= 70:
        return 'adult'
    elif 70 < age <= 100:
        return 'retiree'
    else:
        return 'unknown'

data['AgeGroup'] = data['age'].apply(get_age_group)

print(data[['age', 'AgeGroup']].head(10))
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
data['is_rich'] = (data['salary'] == '>50K').astype(int)

stats = data.groupby('AgeGroup')['is_rich'].agg(['sum', 'mean'])

stats.columns = ['Количество >50K', 'Доля >50K']

print(stats)

most_frequent_group = stats['Доля >50K'].idxmax()
max_rate = stats['Доля >50K'].max()

print(f"\nЧаще всего зарабатывают больше 50K в группе: {most_frequent_group}")
print(f"Доля богатых в этой группе составляет: {max_rate:.1%}")
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
occupation_counts = data['occupation'].value_counts()
occupation_counts

def filter_func(group):

    mean_age_check = group['age'].mean() <= 40

    hours_check = (group['hours-per-week'] > 5).all()
    
    return mean_age_check and hours_check

filtered_data = data.groupby('occupation').filter(filter_func)

print("\nThe professions that passed the filter:")
unique_occupations = filtered_data['occupation'].unique()
print(unique_occupations)

if len(unique_occupations) == 0:
    print("There is no group satisfying both conditions.")
```
