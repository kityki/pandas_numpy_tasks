**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(0, 10, size=(8, 3))
B = np.random.randint(0, 10, size=(2, 2))

print(f"Array A: {A}")
print(f"B array: {B}")

mask_b1 = np.isin(A, B[0]).any(axis=1)
mask_b2 = np.isin(A, B[1]).any(axis=1)

print(f"mask b: {mask_b1}")

final_mask = mask_b1 & mask_b2 

result = A[final_mask]

print(f"The result:{result}")
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
matrix = np.array([[7, 8, 9],
       [3, 2, 2],
       [0, 1, 6],
       [7, 7, 7],
       [7, 9, 0],
       [5, 7, 5],
       [2, 2, 2],
       [8, 9, 6],
       [4, 4, 4],
       [1, 1, 1]])


first_col = matrix[:, [0]]

is_equal_mask = (matrix == first_col)

identical_rows = is_equal_mask.all(axis=1)

result = matrix[~identical_rows]

result
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
matrix = np.array([[7, 8, 9],
       [3, 2, 2],
       [0, 1, 6],
       [7, 7, 7],
       [7, 9, 0],
       [2, 2, 2],
       [2, 2, 2],
       [8, 9, 6],
       [7, 7, 7],
       [1, 1, 5]])

first_row_num = matrix[0:, 0:1]
mask1 = (matrix == first_row_num)
repeated_rows = mask1.all(axis=1)
result = matrix[~repeated_rows]

result



```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
print(f"using NumPy:")

X = np.array([[1, 0, 1], 
              [2, 0, 2], 
              [3, 0, 3], 
              [4, 4, 4]])
d = np.diag(X)
no_zero = d[d != 0]
result = np.prod(no_zero)
print(result)

print(f"using Python lists:")

lst = [[1, 0, 1], 
        [2, 0, 2], 
        [3, 0, 3], 
        [4, 4, 4]]
total = 1
limit = min(len(lst), len(lst[0])) 

for i in range(limit):
    element = lst[i][i] 
    print(element)
    if element != 0:
        print(element)
        total *= element

print(f"Final result: {total}")
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
print(f"The solution using Python lists:")

lst = [6, 2, 0, 3, 0, 0, 5, 7, 0]
new_lst = []
for i in range(len(lst) -1):
    if lst[i] == 0:
         i += 1
         new_lst.append(lst[i])
print(max(new_lst))    

print(f"The solution using NumPy:")

arr = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

mask = (arr[:-1] == 0)
mask1 = arr[1:][mask]
res = np.max(mask1)
print(res) 
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
print(f"The Python lists solution:")

lst = [2, 2, 3, 2, 4, 3, 3, 1, 4, 5]
values = [lst[0]]
counts = [1]

for i in range(1, len(lst)):
    if lst[i] == lst[i - 1]:
        counts[-1] += 1
    else:
        values.append(lst[i])
        counts.append(1)

result = (np.array(values), np.array(counts))

print("Values:", result[0])
print("Counts:", result[1])

print(f"The NumPy solution:")

arr = np.array([2, 2, 3, 2, 4, 3, 3, 1, 4, 5])

unique_arr = np.where(np.diff(arr) != 0)[0] + 1

arr_values = arr[np.insert(unique_arr, 0, 0)]

all_indices = np.concatenate(([0], unique_arr, [len(arr)]))
arr_counts = np.diff(all_indices)

print("Values:", arr_values)
print("Counts:", arr_counts)


```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
print(f"The Python lists solution:")

x = [1, 0, 3, 2]
y = [2, 1, 6, 5]

matrix = []

for val_x in x:
    row = []
    for val_y in y:
        dist = math.sqrt((val_x - val_y)**2)
        row.append(dist)
    matrix.append(row)

for row in matrix:
    print(row)

print(f"The NumPy solution:")

x = np.array([
    [0, 0],
    [1, 1],
    [2, 2]
])

# Y: 2 objects, 2 dimensions
y = np.array([
    [0, 1],
    [1, 0]
])

print(f"Shapes: x={x.shape}, y={y.shape}")

start_my = time.time()

diff = x[:, np.newaxis, :] - y[np.newaxis, :, :]

my_dist_matrix = np.sqrt(np.sum(diff**2, axis=-1))

end_my = time.time()

start_scipy = time.time()

scipy_dist_matrix = cdist(x, y, metric='euclidean')

end_scipy = time.time()

print("-" * 15)

print("My Result Matrix:")
print(my_dist_matrix)

print("-" * 15)

print("SciPy Result Matrix:")
print(scipy_dist_matrix)

match = np.allclose(my_dist_matrix, scipy_dist_matrix)

print("-" * 15)

print(f"Do results match: {match}")

print(f"My Time:    {end_my - start_my:.6f}")
print(f"SciPy Time: {end_scipy - start_scipy:.6f}")
```
